## Accessing Elements in a DataFrame

---



There are two main ways to access elements in a DataFrame:

1. `loc[]`
2. `iloc[]`

### `loc[]`

The `loc[]` method uses **row labels** and **column labels** to access data.

### `iloc[]`

The `iloc[]` method uses **integer-based indices** (row and column positions) to access data.


### Using `loc[]`

````python
coffee.loc[0]
````

````python
coffee.loc[[0,1,6]]
````

````python
coffee.loc[0:6] #Using slicing
````

### Let's use labels to access data
`loc[]` uses the following format `coffee.loc[Rows,Columns]`.

We can also pass a list as an argument with labels

````python
coffee.loc[[0,1,2],["Day","Units Sold"]]
````

### Using `iloc[]`

The `iloc[]` method uses the following syntax:

```python
coffee.iloc[rows, columns]
```

Unlike `loc[]`, which uses row and column labels, `iloc[]` only accepts **integer-based indices** as arguments.

#### Example

```python
coffee.iloc[0, 1]
```

This returns the value located in the first row (`0`) and the second column (`1`).

We can also pass a list of indices to access multiple rows and columns:

```python
coffee.iloc[[0, 2], [1, 2]]
```

This returns the values from rows at indices `0` and `2` and columns at indices `1` and `3`.

````python
coffee.iloc[[0, 2], [1, 2]]
````

````python
coffee.iloc[0:6,1] #Using slicing
````

### Accessing a Single Element

We can use the `at()` and `iat()` methods to access a single element in a DataFrame.

- `at()` is analogous to `loc[]` and uses **row and column labels**.
- `iat()` is analogous to `iloc[]` and uses **integer-based indices**.

Unlike `loc[]` and `iloc[]`, these methods can only access a single element at a time.

#### Examples

Using `at()`:

```python
coffee.at['Monday', 'Coffee Type']
```

Using `iat()`:

```python
coffee.iat[0, 1]
```

Both statements return a single value from the DataFrame.

````python
coffee.at[0,"Day"]
````

````python
coffee.at[0:5,"Day"] # This raises an error because at() can only access a single element, not multiple rows.
````

## Sorting Data

We can use the `sort_values()` method to sort a DataFrame based on one or more columns.

By default, the data is sorted in **ascending order**. To sort in **descending order**, pass `ascending=False` as an argument.

### Example

```python
coffee.sort_values("Units Sold")
```

This sorts the DataFrame by the `Units Sold` column in ascending order.

### Descending Order

```python
coffee.sort_values("Units Sold", ascending=False)
```

This sorts the DataFrame by the `Units Sold` column in descending order.

````python
coffee.sort_values("Units Sold")
````

### Sorting by Multiple Columns

We can sort a DataFrame using multiple columns by passing a list of column names to the `by` parameter.

```python
coffee.sort_values(
    by=["Coffee Type", "Units Sold"],
    ascending=[True, False]
)
```

In this example:

- `Coffee Type` is sorted in ascending order.
- `Units Sold` is sorted in descending order within each coffee type.

---

### Sorting and Slicing

We can combine sorting with slicing to view only a subset of the sorted data.

```python
coffee.sort_values(
    "Units Sold",
    ascending=False
)
```

This returns the top 5 rows with the highest values in the `Units Sold` column.

---

### Sorting by Multiple Columns and Slicing

```python
coffee.sort_values(
    ["Coffee Type", "Units Sold"],
    ascending=[True, False]
)
```

This first sorts the DataFrame by `Coffee Type` and `Units Sold`, then returns the first 10 rows of the sorted DataFrame.

````python
coffee.sort_values(
    by=["Coffee Type", "Units Sold"]
)
````

````python
coffee.sort_values(
    ["Coffee Type", "Units Sold"],
    ascending=[True, False]
)
````

## Iterating Through Rows

We can use the `iterrows()` method to iterate through the rows of a DataFrame.

The method returns:
- The row index.
- The row data as a Pandas Series.

### Example

```python
for index, row in coffee.iterrows():
    print(index)
    print(row)
    print("\n")
```

In the above example:

- `index` stores the index label of the current row.
- `row` stores the row data as a Series.
- The loop iterates through each row of the DataFrame one at a time.

### Accessing Individual Values

We can access values from the current row using column labels.

```python
for index, row in coffee.iterrows():
    print(row["Coffee Type"])
    print(row["Units Sold"])
```

This prints the values of the `Coffee Type` and `Units Sold` columns for each row.

````python
for index,row in coffee.iterrows():
  print(index)
  print(row)
  print("\n\n")
````

````python
for index, row in coffee.iterrows():
    print(row["Coffee Type"])
    print(row["Units Sold"])
````
