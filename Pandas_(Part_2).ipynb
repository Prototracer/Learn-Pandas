#Filtering Data

We can pass Boolean conditions to the `loc[]` method to filter rows in a DataFrame.

The condition generates a Boolean Series containing `True` and `False` values. Rows corresponding to `True` are selected, while rows corresponding to `False` are excluded.


````python
bios.loc[bios['height_cm']>180]
````

````python
bios.loc[bios['height_cm']>210,['name','height_cm']]
````

### Using Multiple Conditions with `&` and `|`

We can combine multiple Boolean conditions using the following operators:

- `&` → Logical AND
- `|` → Logical OR

When using multiple conditions, each condition must be enclosed in parentheses.

### Using `&` (AND)

```python
coffee.loc[
    (coffee["Units Sold"] > 50) &
    (coffee["Coffee Type"] == "Latte")
]
```

This returns rows where:

- `Units Sold` is greater than 50, **and**
- `Coffee Type` is `"Latte"`.

### Using `|` (OR)

```python
coffee.loc[
    (coffee["Units Sold"] > 50) |
    (coffee["Coffee Type"] == "Latte")
]
```

This returns rows where:

- `Units Sold` is greater than 50, **or**
- `Coffee Type` is `"Latte"`.

### Combining Multiple Conditions

```python
coffee.loc[
    ((coffee["Units Sold"] > 50) &
     (coffee["Coffee Type"] == "Latte")) |
    (coffee["Price Per Unit"] > 4)
]
```

This returns rows where:

- Both `Units Sold > 50` and `Coffee Type == "Latte"`, **or**
- `Price Per Unit > 4`.

> **Note:** Use `&` and `|` for DataFrame filtering instead of Python's `and` and `or`. Also, remember to wrap each condition in parentheses.

````python
bios.loc[(bios['height_cm']>210) & (bios['born_country']=='USA')]
````

## String Operations with `str`

Pandas provides the `str` accessor to perform string operations on a Series.

| `str.upper()` | Converts text to uppercase |
| `str.title()` | Converts text to title case |

| `str.startswith()` | Checks if a string starts with a pattern |
| `str.endswith()` | Checks if a string ends with a pattern |

## Using `str.contains()`

The `str.contains()` method checks whether each string in a column contains a specified pattern.

It returns a Boolean Series containing `True` and `False` values.

### Syntax

```python
Series.str.contains(pattern)
```

### Example

```python
coffee["Coffee Type"].str.contains("Latte")
```

Output:

```python
0     True
1    False
2     True
3    False
dtype: bool
```

---

### Filtering Rows with `str.contains()`

Since `str.contains()` returns a Boolean Series, it can be used with `loc[]` to filter rows.

```python
coffee.loc[
    coffee["Coffee Type"].str.contains("Latte")
]
```

This returns all rows where the `Coffee Type` column contains the word `"Latte"`.

---

### Case-Insensitive Search

```python
coffee.loc[
    coffee["Coffee Type"].str.contains("latte", case=False)
]
```

This matches `"Latte"`, `"LATTE"`, `"latte"`, and other case variations.

---

### Matching Multiple Patterns

```python
coffee.loc[
    coffee["Coffee Type"].str.contains("Latte|Espresso")
]
```

This returns rows where the `Coffee Type` contains either `"Latte"` or `"Espresso"`.

> **Note:** Most string operations in Pandas are performed using the `str` accessor. It allows string methods to be applied efficiently to an entire column at once.

## Using `isin()`

The `isin()` method is used to check whether each value in a column exists in a specified list of values.

It returns a Boolean Series containing `True` and `False` values.

### Syntax

```python
Series.isin(values)
```

### Example

```python
coffee["Coffee Type"].isin(["Latte", "Espresso"])
```

Output:

```python
0     True
1    False
2     True
3    False
dtype: bool
```

This checks whether each value in the `Coffee Type` column is either `"Latte"` or `"Espresso"`.

---

### Filtering Rows with `isin()`

Since `isin()` returns a Boolean Series, it can be used with `loc[]` to filter rows.

```python
coffee.loc[
    coffee["Coffee Type"].isin(["Latte", "Espresso"])
]
```

This returns all rows where the `Coffee Type` is either `"Latte"` or `"Espresso"`.

---

### Using `isin()` with Multiple Values

```python
coffee.loc[
    coffee["Day"].isin(["Monday", "Wednesday", "Friday"])
]
```

This returns all rows where the `Day` is Monday, Wednesday, or Friday.

---

### Excluding Values

We can use the `~` operator to select rows that do **not** match the specified values.

```python
coffee.loc[
    ~coffee["Coffee Type"].isin(["Latte", "Espresso"])
]
```

This returns all rows where the `Coffee Type` is neither `"Latte"` nor `"Espresso"`.

> **Note:** `isin()` is often more readable and efficient than combining multiple equality conditions with the `|` (OR) operator.

````python
bios[bios.born_country.isin(['USA','FRA'])]
````

## Understanding the `axis` Parameter

Many Pandas methods use the `axis` parameter to specify whether an operation should be performed on rows or columns.

- `axis=0` → Rows (operate down the rows)
- `axis=1` → Columns (operate across the columns)

### Example: Dropping Columns

```python
coffee.drop("price", axis=1)
```

This removes the `price` column because `axis=1` refers to columns.

### Example: Dropping Rows

```python
coffee.drop(0, axis=0)
```

This removes the row with index `0` because `axis=0` refers to rows.

### Easy Way to Remember

- `axis=0` → Vertical direction ↓ (Rows)
- `axis=1` → Horizontal direction → (Columns)

> **Note:** Many Pandas methods now provide more explicit parameters such as `columns=` and `index=`, which are often easier to read than using `axis`.

````python
coffee.drop("Day",axis=1)
````

### `rolling()` Method

The `rolling()` method is used to perform calculations over a moving window of values.

```python
coffee["3-Day Avg Sales"] = coffee["Units Sold"].rolling(3).mean()

coffee.head()
```

This calculates the 3-row moving average of the `Units Sold` column.
### Understanding a Moving Window

A moving window is a fixed-size group of consecutive rows that moves through the data one row at a time.

For example, if `Units Sold` contains:

```text
25
15
30
20
35
```

Using:

```python
coffee["Units Sold"].rolling(3).mean()
```

Pandas first takes the first 3 values (`25, 15, 30`) and calculates their average. Then the window moves down by one row and uses (`15, 30, 20`), and so on.

```text
Window 1: [25, 15, 30] → 23.33
Window 2: [15, 30, 20] → 21.67
Window 3: [30, 20, 35] → 28.33
```

This technique is commonly used to analyze trends and smooth out fluctuations in data.

````python
coffee["3-Day Avg Sales"] = coffee["Units Sold"].rolling(3).mean()
coffee.head()
````
